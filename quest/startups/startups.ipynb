```python
startups=startups.rename(columns={'Valuation_US$_billions':'Valuation_US_billions'})
startups
```

```python
#counting the amount of startups within normal distribution to exclude the outl
IQL=startups['Valuation_US_billions'].quantile(0.75)-startups['Valuation_US_billions'].quantile(0.25)

if startups['Valuation_US_billions'].quantile(0.25)-1.5*IQL >=0:
    min_val=startups['Valuation_US_billions'].quantile(0.25)-1.5*IQL
else:
    min_val=0
    
max_val=startups['Valuation_US_billions'].quantile(0.75)+3*IQL

print('IQL=',IQL)
print('min_val=',min_val)
print('max_val=',max_val)
```

```python
count=0
for i in startups['Valuation_US_billions']:
    if i>=min_val and i<=max_val:
        count+=1
print(count)
```

```python
#checking outl
df_outliners_max=startups.loc[(startups['Valuation_US_billions'] > max_val)]
df_outliners_max

df_outliners_min=startups.loc[(startups['Valuation_US_billions'] <1)]
print(df_outliners_min)
```

```python
# getting final df

startups_cleaned=startups.loc[(startups['Valuation_US_billions']>=min_val) & (startups['Valuation_US_billions']<=max_val)]
startups_cleaned
```

```python
#saving CSV for SQL for further work
path="C:\\Users\\user\\Desktop\\ironhack_study\\quest\\startups\\tables\\startups_cleaned.csv"
startups_cleaned.to_csv(path, index=True)
```
