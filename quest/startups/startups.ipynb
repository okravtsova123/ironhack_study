```python
startups=startups.rename(columns={'Valuation_US$_billions':'Valuation_US_billions'})
startups
```

```python
#counting the amount of startups within normal distribution to exclude the outl
IQL=startups['Valuation_US_billions'].quantile(0.75)-startups['Valuation_US_billions'].quantile(0.25)

if startups['Valuation_US_billions'].quantile(0.25)-1.5*IQL >=0:
    min_val=startups['Valuation_US_billions'].quantile(0.25)-1.5*IQL
else:
    min_val=0
    
max_val=startups['Valuation_US_billions'].quantile(0.75)+3*IQL

print('IQL=',IQL)
print('min_val=',min_val)
print('max_val=',max_val)
```

```python
count=0
for i in startups['Valuation_US_billions']:
    if i>=min_val and i<=max_val:
        count+=1
print(count)
```

```python
#checking outl
df_outliners_max=startups.loc[(startups['Valuation_US_billions'] > max_val)]
df_outliners_max

df_outliners_min=startups.loc[(startups['Valuation_US_billions'] <1)]
print(df_outliners_min)
```

```python
# getting final df

startups_cleaned=startups.loc[(startups['Valuation_US_billions']>=min_val) & (startups['Valuation_US_billions']<=max_val)]
startups_cleaned
```

```python
#saving CSV for SQL for further work
path="C:\\Users\\user\\Desktop\\ironhack_study\\quest\\startups\\tables\\startups_cleaned.csv"
startups_cleaned.to_csv(path, index=True)
```

```python
#checning the correlation between factors
startups_gr=pd.read_csv('https://raw.githubusercontent.com/okravtsova123/ironhack_study/main/quest/startups/tables/startups_with%20size_group_12.csv')
startups_gr

#updated table from sql for countries
countries=pd.read_csv('https://raw.githubusercontent.com/okravtsova123/ironhack_study/main/quest/startups/tables/countries_from_sql_for%20correlation.csv')

#concatingating with other table
full_table=pd.merge(startups_gr, countries, on='Country')
full_table

full_table.isnull().sum()
#full_table.count()

#cleaning nulls with "not known" for industries
full_table=full_table.fillna('not known')

full_table.isnull().sum()
full_table.nunique()
```

```python
full_table
```

```python
#calclating correlations
correlation_rule_of_law = full_table['Valuation_US_billions'].corr(full_table['rule_of_law'])
print('correlation_rule_of_law:', correlation_rule_of_law)

correlation_gdp = full_table['Valuation_US_billions'].corr(full_table['GDP_blnUSD'])
print('correlation_gdp :', correlation_gdp)

correlation_gdp_p_c = full_table['Valuation_US_billions'].corr(full_table['GDP_per_capita_tUSD'])
print('correlation_gdp_p_c:', correlation_gdp_p_c)

correlation_population = full_table['Valuation_US_billions'].corr(full_table['Population'])
print('correlation_population:', correlation_population)

correlation_inn_score = full_table['Valuation_US_billions'].corr(full_table['Innovation_Score'])
print('correlation_inn_score:', correlation_inn_score)

correlation_tax_score = full_table['Valuation_US_billions'].corr(full_table['Tax_Score'])
print('correlation_tax_score:', correlation_tax_score)
```

```python
df = full_table.loc[full_table['Tax_Score'].to_numpy() != 0]
correlation_tax_score = df['Valuation_US_billions'].corr(df['Tax_Score'].astype(float))
correlation_tax_score
```
