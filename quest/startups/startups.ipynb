```python
df1=df1.rename(columns={'Valuation_US$_billions':'Valuation_US_billions'})
```

```python
#counting the amount of statups within normal distribution to exclude the outl
IQL=df1['Valuation_US_billions'].quantile(0.75)-df1['Valuation_US_billions'].quantile(0.25)

if df1['Valuation_US_billions'].quantile(0.25)-1.5*IQL >=0:
    min_val=df1['Valuation_US_billions'].quantile(0.25)-1.5*IQL
else:
    min_val=0
    
max_val=df1['Valuation_US_billions'].quantile(0.75)+3*IQL

print('IQL=',IQL)
print('min_val=',min_val)
print('max_val=',max_val)
```

```python
count=0
for i in df1['Valuation_US_billions']:
    if i>=min_val and i<=max_val:
        count+=1
print(count)
```

```python
#checking outl
df_outliners_max=df1.loc[(df1['Valuation_US_billions'] > max_val)]
df_outliners_max

df_outliners_min=df1.loc[(df1['Valuation_US_billions'] <1)]
print(df_outliners_min)
```

```python
# getting final df

df_final=df1.loc[(df1['Valuation_US_billions']>=min_val) & (df1['Valuation_US_billions']<=max_val)]
df_final
```

```python
df_final.to_csv('startups_cleaned.csv')
```
