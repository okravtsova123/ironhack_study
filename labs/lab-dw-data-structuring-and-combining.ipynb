```python
import pandas as pd
df1=pd.read_csv('https://raw.githubusercontent.com/data-bootcamp-v4/data/main/file1.csv')
df2=pd.read_csv('https://raw.githubusercontent.com/data-bootcamp-v4/data/main/file2.csv')
df3=pd.read_csv('https://raw.githubusercontent.com/data-bootcamp-v4/data/main/file3.csv')
```

```python
#cleaning the df1

df1.columns=[name.lower() for name in df1.columns]
df1.columns=[name.replace(" ","_") for name in df1.columns]
df1.rename(columns= {'st': 'state'}, inplace=True)

def gender_cleaning(x):
    if x in ("F","Femal","Female","female"):
        x="F"
    elif x in ("M","Male","Mal"):
        x="M"
    return x

df1["gender"]=df1["gender"].apply(gender_cleaning)

def state_cleaning(state):
    if state=="AZ":
        state="Arizona"
    elif state=="WA":
        state="Washington"
    elif state=="Cali":
        state="California"
    return state

df1["state"]=df1["state"].apply(state_cleaning)

df1["education"]=["Bachelor" if i=="Bachelors" else i for i in df1["education"]]

df1['customer_lifetime_value']=[ltv if type(ltv)==float else ltv.replace("%","") for ltv in df1['customer_lifetime_value']]

df1['vehicle_class']=["Luxury" if car in ["Sports Car", "Luxury SUV","Luxury Car"] else car for car in df1['vehicle_class']]

df1["customer_lifetime_value"] = df1["customer_lifetime_value"].astype(float)

df1["number_of_open_complaints"]=[nooc.split("/")[1] if type(nooc)!=float else nooc for nooc in df1["number_of_open_complaints"]]

df1["number_of_open_complaints"]=df1["number_of_open_complaints"].astype(float)

df1_new=df1.dropna(subset="customer")

df1_new["customer_lifetime_value"]=df1_new["customer_lifetime_value"].fillna(df1_new["customer_lifetime_value"].median())

gender_mode= df1_new["gender"].mode()

df1_new["gender"]=df1_new["gender"].fillna(gender_mode.item())
```

```python
#cleaned the 2nd
df2.columns=[name.lower() for name in df2.columns]
df2.columns=[name.replace(" ","_") for name in df2.columns]
df2.rename(columns= {'st': 'state'}, inplace=True)

df2["gender"]=df2["gender"].apply(gender_cleaning)
df2["state"]=df2["state"].apply(state_cleaning)

df2["education"]=["Bachelor" if i=="Bachelors" else i for i in df2["education"]]
df2['customer_lifetime_value']=[ltv if type(ltv)==float else ltv.replace("%","") for ltv in df2['customer_lifetime_value']]
df2['vehicle_class']=["Luxury" if car in ["Sports Car", "Luxury SUV","Luxury Car"] else car for car in df2['vehicle_class']]
df2["customer_lifetime_value"] = df2["customer_lifetime_value"].astype(float)
df2["number_of_open_complaints"]=[nooc.split("/")[1] if type(nooc)!=float else nooc for nooc in df2["number_of_open_complaints"]]
df2["number_of_open_complaints"]=df1["number_of_open_complaints"].astype(float)

df2_new=df2.dropna(subset="customer")
df2_new["customer_lifetime_value"]=df2_new["customer_lifetime_value"].fillna(df2_new["customer_lifetime_value"].median())
gender_mode= df2_new["gender"].mode()
df2_new["gender"]=df2_new["gender"].fillna(gender_mode.item())
```

```python
#cleaned the 3rd
df3.columns=[name.lower() for name in df3.columns]
df3.columns=[name.replace(" ","_") for name in df3.columns]
df3.rename(columns= {'st': 'state'}, inplace=True)

df3["gender"]=df3["gender"].apply(gender_cleaning)
df3["state"]=df3["state"].apply(state_cleaning)

df3["education"]=["Bachelor" if i=="Bachelors" else i for i in df3["education"]]
df3['customer_lifetime_value']=[ltv if type(ltv)==float else ltv.replace("%","") for ltv in df3['customer_lifetime_value']]
df3['vehicle_class']=["Luxury" if car in ["Sports Car", "Luxury SUV","Luxury Car"] else car for car in df3['vehicle_class']]
df3["customer_lifetime_value"] = df3["customer_lifetime_value"].astype(float)
df3["number_of_open_complaints"]=df3["number_of_open_complaints"].astype(float)

df3_new=df3.dropna(subset="customer")
df3_new["customer_lifetime_value"]=df3_new["customer_lifetime_value"].fillna(df3_new["customer_lifetime_value"].median())
gender_mode= df3_new["gender"].mode()
df3_new["gender"]=df3_new["gender"].fillna(gender_mode.item())
```

```python
df_merged=pd.concat([df1_new,df2_new,df3_new])
```
